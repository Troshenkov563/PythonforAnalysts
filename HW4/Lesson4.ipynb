#### 1.1
Данные из прошлого дз

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df.head()
```

#### 1.2
Изучите стоимости недвижимости

```python
plt.figure(figsize=(15, 5))
sns.histplot(df['price'])
plt.ticklabel_format(style='plain')
plt.grid()
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.title('Распределение стоимости недвижимости');
```

Стоимость домов находится в диапазоне от 200 тысяч - до 1 млн.

#### 1.3
Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(15, 5))
sns.histplot(df['sqft_living'], bins=150)
plt.ticklabel_format(style='plain')
plt.grid()
plt.title('Распределение квадратуры жилой')
plt.xlabel('Кв. метры')
plt.ylabel('Количество');
```

Средняя жилая площадь ~ 1800 кв.

#### 1.4
Изучите распределение года постройки

```python
plt.figure(figsize=(15, 5))
sns.histplot(df['yr_built'], bins=15)
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

Самые высокие темпы строительства были в 2000-x гг.

## Задание 2:

#### 2.1
Постройте график, сделайте выводы.

```python
data = df['waterfront'].value_counts()
plt.pie(data.values, labels=data.index, autopct='%1.1f%%');
plt.legend(['NO', 'YES']);
```

Меньше 1% домов имеют вид на набережную.

#### 2.2
Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
plt.pie(data.values, labels=data.index, autopct='%1.1f%%');
plt.legend(data.index);
```

Самые распространенные - одноэтажные дома, за ними идут дома с двумя этажами. Дома с мансардой и выше двух этажей составляют малую часть от всех построек.

#### 2.3
Изучите распределение состояния домов

```python
plt.figure(figsize=(6, 4))
sns.barplot(x = data.index, y = data)
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Кол-во');
```

## Задание 3:

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.boxplot(x=df['price'], y=df['condition'].astype('str'), whis=1.5, showfliers=False)
plt.title('Зависимость цены недвижимости от состояния');
```

Цена недвижимости коррелируется с состоянием дома. Чем выше цена, тем выше состояние дома.

```python
sns.boxplot(x=df['yr_built'], y=df['condition'].astype('str'), whis=1.5, showfliers=False)
plt.title('Зависимость состояния недвижимости от года постройки');
```

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5, showfliers=False)
plt.title('Зависимость цены недвижимости от наличия красивого вида');
```

Дома с красивым видом стоят дороже.
